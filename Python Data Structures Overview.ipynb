# how to mutate the values of tuples<br>
## Convert Tuples into List --> Apply Transformations --> Convert Lists into Tuples

```python
aTuple = ('275', '54000', '0.0', '5000.0', '0.0')
print("Before transformation: ",aTuple)
lst = list(aTuple)
lst[0] = '300'
t = tuple(lst)
print("After transformation: ",t)
```

## String Mutation

```python
def split(word): 
    return [char for char in word]  
      
# Driver code 
Lis_2 = split(string)
print(Lis_2)
```
